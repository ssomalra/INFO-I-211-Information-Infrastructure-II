# Question 2
Your local gym is helping a nearby elementary school by providing diet and workout data for their various patrons. Import the exercise data from Seaborn. Provide both the figures (in PNG format) and code for creating the following plots. 

```python
exercise = sns.load_dataset('exercise')
exercise
```

### Part A
Create a heatmap for the participants, workout time, and pulse. (Hint use the pivot command from Pandas to get the data in appropriate format)

```python
graph = exercise.pivot(columns='time', index='id', values='pulse')
graph
```

```python
sns.heatmap(graph).set(title='Heatmap for Participants, Workout Time, and Pulse')
```

### Part B 
Create a categorical plot for pulse by diet and by type of exercise

```python
sns.swarmplot(x='kind', y='pulse', hue='diet', data=exercise).set(title='Distribution of Pulse by Diet and Type of Exercise')
```
